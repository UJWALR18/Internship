# Q1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) View

```python
import pandas as pd
import selenium
from bs4 import BeautifulSoup
import time
from selenium import webdriver
import requests
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
name=[]
try:
    n=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in n:
        name.append(i.text)
except NoSuchElementException:
            name.append('No details available')
except StaleElementReferenceException:
    name.append('No details available')
```

```python
Rank=[]
try:
    r=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in r:
        Rank.append(i.text)
except NoSuchElementException:
            Rank.append('No details available')
except StaleElementReferenceException:
    Rank.append('No details available')
```

```python
artist=[]
try:
    a=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in a:
        artist.append(i.text)
except NoSuchElementException:
            artist.append('No details available')
except StaleElementReferenceException:
    artist.append('No details available')
```

```python
date=[]
try:
    d=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in d:
        date.append(i.text)
except NoSuchElementException:
            date.append('No details available')
except StaleElementReferenceException:
    date.append('No details available')
```

```python
view=[]
try:
    v=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in v:
        view.append(i.text)
except NoSuchElementException:
            view.append('No details available')
except StaleElementReferenceException:
    view.append('No details available')
```

```python
df = pd.DataFrame()
df['Rank']=Rank 
df['Name']=name
df['Artist']=artist
df['Views']=view
df['Upload_Date']=date
df
```

# Q-7. Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/
You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes


```python
#loading website
driver=webdriver.Chrome() # creating automated chrome window
driver.maximize_window() # maximizing
driver.get("https://www.imdb.com/list/ls095964455/") # going to website
```

```python
# creating empty disctionary
imdb_series = {}
imdb_series['Name'] = []
imdb_series['YearSpan'] = []
imdb_series['Genre'] = []
imdb_series['RunTime'] = []
imdb_series['Rating'] = []
imdb_series['Votes'] = []
```

```python
#Scraping the above data from IMDB Website
names = driver.find_elements(By.XPATH,"//h3[@class='lister-item-header']/a")
for i in names:
    if i.text is None:
        imdb_series['Name'].append("-")
    else:
        imdb_series['Name'].append(i.text)
        
yearSpan = driver.find_elements(By.XPATH,"//h3[@class='lister-item-header']/span[2]")
for i in yearSpan:
    if i.text is None:
        imdb_series['YearSpan'].append("-")
    else:
        imdb_series['YearSpan'].append(i.text.replace('(','').replace(')',''))

        
genre = driver.find_elements(By.XPATH,"//p[@class='text-muted text-small']/span[5]")
for i in genre:
    if i.text is None:
        imdb_series['Genre'].append("-")
    else:
        imdb_series['Genre'].append(i.text)
        
runtime = driver.find_elements(By.XPATH,"//p[@class='text-muted text-small']/span[3]")
for i in runtime:
    if i.text is None:
        imdb_series['RunTime'].append("-")
    else:
        imdb_series['RunTime'].append(i.text)
        
        
rating = driver.find_elements(By.XPATH,"//div[@class ='ipl-rating-star small']/span[2]")
for i in rating:
    if i.text is None:
        imdb_series['Rating'].append("-")
    else:
        imdb_series['Rating'].append(i.text)

votes = driver.find_elements(By.XPATH,"//div[@class='lister-item-content']/p[4]/span[2]")
for i in votes:
    if i.text is None:
        imdb_series['Votes'].append("-")
    else:
        imdb_series['Votes'].append(i.text)        
```

```python
    
# Creating dataframe using pandas lib
IMDB_Series = pd.DataFrame(imdb_series)

IMDB_Series
```

# Q8. Details of Datasetsfrom UCI machine learning repositories.
Url = https://archive.ics.uci.edu/
You have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute
G) Year

```python
#loading website
driver=webdriver.Chrome() # creating automated chrome window
driver.get("https://archive.ics.uci.edu/") # going to website
```

```python

# click on to all view all dataset
driver.find_element(By.XPATH,'/html/body/table[2]/tbody/tr/td/span/b/a').click()
```

```python
#scrapping weblink
weblinks=[]

#scraping weblinks
for i in driver.find_elements(By.XPATH,'//p[@class="normal"]/b/a'):
    weblinks.append(i.get_attribute('href'))
```

```python
#scrapping information 
# creating empty lists
Name=[]
Type=[]
Task=[]
Att_Type=[]
Instance=[]
Attributes=[]
Year=[]
```

```python
for i in weblinks:
    try:
        driver.get(i)
    # Scrapping name of dataset
        name=driver.find_element(By.XPATH,'//span[@class="heading"]')
        Name.append(name.text)

    # Scrapping Type of dataset
        tipe=driver.find_element(By.XPATH,'(//p[@class="normal"])[3]')
        Type.append(tipe.text)

    # Scrapping Task
        task=driver.find_element(By.XPATH,'(//p[@class="normal"])[15]')
        Task.append(task.text)
    
    # Scrapping Attribute type
        att=driver.find_element(By.XPATH,'(//p[@class="normal"])[9]')
        Att_Type.append(att.text)

    # Scrapping Instance
        instance=driver.find_element(By.XPATH,'(//p[@class="normal"])[5]')
        Instance.append(instance.text)

    # Scrapping Attributes
        attributes=driver.find_element(By.XPATH,'(//p[@class="normal"])[11]')
        Attributes.append(attributes.text)

    # Scrapping Year
        year=driver.find_element(By.XPATH,'(//p[@class="normal"])[13]')
        Year.append(year.text)
    
    except:
        pass
```

```python
# making a dataframe
dataset=pd.DataFrame({'Name of Dataset':Name,'Type of Dataset':Type,'Task':Task,'Attribute type':Att_Type,'No. of Instances':Instance,'No.of Attributes':Attributes,'Year':Year})

dataset
```

```python
#loading website
driver=webdriver.Chrome() # creating automated chrome window
driver.get("https://www.naukri.com/hr-recruiters-consultants") # going to website
```

```python
# sending keys "Data Science" in search box
driver.find_element(By.XPATH,'//input[@class="sugInp"]').send_keys('Data Science')
#clicking on search button
driver.find_element(By.XPATH,'//button[@class="fl qsbSrch blueBtn"]').click()
```

```python
#scrapping the required information
# creating empty lists
Name=[]
Designation=[]
Company_Name=[]
Skills=[]
Location=[]
```

```python
#scrapping Recruiter's name
recruit=driver.find_elements(By.XPATH,'//span[@class="fl ellipsis"]')
for i in recruit:
    Name.append(i.text)

#scrapping Recruiter's designation
desg=driver.find_elements(By.XPATH,'//span[@class="ellipsis clr"]')
for i in desg:
    Designation.append(i.text)

#scrapping Recruiter's Company
Company=[]
company=driver.find_elements(By.XPATH,'//a[@class="ellipsis"]')
for i in company:
    Company.append(i.text)
for i in range(1,100,2):
    Company_Name.append(Company[i])

#scrapping skills they hire for
skills=driver.find_elements(By.XPATH,'//div[@class="hireSec highlightable"]')
for i in skills:
    Skills.append(i.text)

#scrapping Company Location
location=driver.find_elements(By.XPATH,'//div[@class="recInfo"]/div/p/span[2]/small')
try:
    for i in location:
        Location.append(i.text)
except NoSuchElementException:
    Location.append('Not Mentioned')

# making a dataframe
Recruiter_data=pd.DataFrame({'Name of Recruiter':Name[:49],'Designation of Recruiter':Designation[:49],'Name of Company':Company_Name[:49],'Skills They Hire for':Skills[:49],'Location':Location[:49]})

Recruiter_data
```
