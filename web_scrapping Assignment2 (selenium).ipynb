# a python program to scrape data for Top 1000 Quotes of All Time

```python

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
#Lets first connect to the driver
driver = webdriver.Chrome(r"chromedriver.exe")
```

```python
driver.get('https://www.azquotes.com/')
```

```python
topquotes_button = driver.find_element(By.XPATH, '//a[@href="/top_quotes.html"]')
topquotes_button.click()
```

```python
Quote=[]
Author=[]
Type_Of_Quotes=[]
```

```python
for page in range(0,10):
    #scrapping quotes from websites
    quotes_tag = driver.find_elements(By.XPATH,'.//a[@class="title"]')
    for i in quotes_tag:
        Quote.append(i.text)
        
        
    #scrapping author
    author_tag = driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in author_tag:
        Author.append(i.text)
        
        
    #scrapping types_of_quotes
    types_of_quotes_tag = driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in types_of_quotes_tag:
        Type_Of_Quotes.append(i.text)
```

```python
print(len(Quote),len(Author),len(Type_Of_Quotes))
```

```python
df=pd.DataFrame({'Quote':Quote,'Author':Author,'Type_of_quotes':Type_Of_Quotes})
df
```
